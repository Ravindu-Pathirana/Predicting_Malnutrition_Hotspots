PCA

```python
index_norm = df[['Overweight_norm', 'Stunting_norm']]
scalar = StandardScaler()
index_norm_scaled = scalar.fit_transform(index_norm)
 
```

```python
index_norm_scaled.shape
```

```python
index_norm_scaled.min()
```

```python
pca = PCA(n_components=1)
index_norm_pca = pca.fit_transform(index_norm_scaled)    
```

```python
print(index_norm_pca)
```

```python
index_norm_pca.shape
```

```python
weights = pca.components_[0]
print(weights)
print(f"Stunting weight: {weights[0]:.4f}")
print(f"Overweight weight: {weights[1]:.4f}")
```

Regenerate Malnutrition Index

```python
df.shape
```

```python
df['Malnutrition_raw'] = (df['Stunting_norm'] + df['Overweight_norm'])
```

```python
min_val = df['Malnutrition_raw'].min()
max_val = df['Malnutrition_raw'].max()

df['Malnutrition_index_sum_norm'] = (
    (df['Malnutrition_raw'] - min_val) / (max_val - min_val)
)
```
