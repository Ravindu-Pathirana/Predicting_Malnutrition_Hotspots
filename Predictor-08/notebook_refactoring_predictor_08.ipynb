```python
len(country_codes)
```

```python
df_filtered = df[df['Country Code'].isin(country_codes)]
```

```python
years = [str(year) for year in range(2000,2024)]
df_final = df_filtered[['Country Name','Country Code'] + years]
```

```python
df_final = df_final.sort_values(by='Country Name').reset_index(drop=True)
```

```python
df_final.head()
```

```python
df_final.size
```

```python
df_final.shape
```

```python
for year in years:
    temp_df = df_final[['Country Name', 'Country Code', year]].copy()
    
    # Rename column to something clean
    temp_df = temp_df.rename(columns={year: 'Prevalance of Anemia %'})

    # ADD year column 
    temp_df['Year'] = int(year)

    # Reorder columns
    temp_df = temp_df[['Country Name', 'Country Code', 'Year', 'Prevalance of Anemia %']]
    
    # Save file
    temp_df.to_csv(f"{year}_maternal_anemia.csv", index=False)
```

```python
df_final.to_csv("Cleaned_Maternal_Anemia_2000_2023.csv", index=False)
```

Find Missing values for each country

```python
missing_count = df_final[years].isnull().sum(axis=1)

df_final_missing_summary = df_final[['Country Name', 'Country Code']].copy()
df_final_missing_summary['Missing Count'] = missing_count

# Sort by highest missing
df_final_missing_summary = df_final_missing_summary.sort_values(by='Missing Count', ascending=False)

print(df_final_missing_summary)
```

```python
missing_years = df_final[years].isnull()

df_final['Missing Years'] = missing_years.apply(
    lambda row: [year for year in years if row[year]], axis=1
)

result = df_final[['Country Name', 'Country Code', 'Missing Years']]

print(result)
```

```python
df_final['Missing Count'] = df_final[years].isnull().sum(axis=1)

df_final['Missing Years'] = df_final[years].isnull().apply(
    lambda row: [year for year in years if row[year]], axis=1
)
df_missing_only = df_final[df_final['Missing Count'] > 0]
final_missing = df_missing_only[['Country Name', 'Country Code', 'Missing Count', 'Missing Years']]

print(final_missing)
```
