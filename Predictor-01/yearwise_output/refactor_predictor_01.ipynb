```python
import pandas as pd
```

```python
years = [str(year) for year in range(2000,2024)]
```

```python
ref_df = pd.read_csv("country_list.csv") 
```

```python
ref_df.head()
```

```python
for year in years:
    df = pd.read_csv(f"food_supply_{year}.csv")
    df.rename(columns={'country': 'Country Name'}, inplace=True)
    df = pd.merge(ref_df, df, on='Country Name', how='left')
    df = df.sort_values(by='Country Name')
    df.to_csv(f"../{year}_food_supply.csv", index=False)
```
