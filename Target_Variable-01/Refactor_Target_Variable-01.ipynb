```python
import pandas as pd
```

```python
ref_df = pd.read_csv("country_list.csv")
df = pd.read_csv("data_stunting.csv")
```

```python
df = df[[
    'SpatialDimValueCode',   # Country Code
    'Location',              # Country Name
    'Period',                # Year
    'Dim1',                  # Sex
    'FactValueNumeric'       # Value (numeric)
]]
```

```python
df = df.rename(columns={
    'SpatialDimValueCode': 'Country Code',
    'Location': 'Country Name',
    'Period': 'Year',
    'FactValueNumeric': 'Value'
})
```

```python
print("Available categories in Dim1:", df['Dim1'].unique())
```

```python
df = df[df['Dim1'] == 'Both sexes']
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df.head()
```

```python
df.shape
```

```python
years = sorted(df['Year'].unique())

for year in years:
    
    # Filter that year
    temp = df[df['Year'] == year][['Country Code', 'Value']]
    
    # Merge with reference (ensures order + missing countries)
    merged = pd.merge(ref_df, temp, on='Country Code', how='left')
    
    # Add year column
    merged['Year'] = year
    
    # Reorder columns
    merged = merged[['Country Name', 'Country Code', 'Year', 'Value']]
    
    # Save file
    merged.to_csv(f"stunting_yearly/{year}_stunting.csv", index=False)
```
